# Broadacre regions

The Australian Bureau of Agricultural and Resource Economics and Sciences (ABARES)
defines a set of Australian broadacre zones and regions.

This notebook gives a basic demonstration of the shapefiles found at:  
`/g/data/ia39/aus-ref-clim-data-nci/shapefiles/shapefiles_broadacre_regions/`

The shapefiles were downloaded from:
https://www.agriculture.gov.au/abares/research-topics/surveys/farm-survey-data

```python
import geopandas as gp
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
#https://apps.agriculture.gov.au/agsurf/regions.html
Image(filename='figures/broadacre_regions.jpg', width=500)
```

```python
broadacre_file = 'shapefiles_broadacre_regions/aagis_asgs16v1_g5a.shp'
```

```python
broadacre_regions = gp.read_file(broadacre_file)
broadacre_regions
```

```python
broadacre_regions.plot(
    figsize=[5, 5],
    column='zone',
    legend=True,
    linewidth=0.5,
    edgecolor='black',
)
plt.title('Broadacre Regions')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
