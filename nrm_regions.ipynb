## NRM regions

The most recent climate projections for Australia make extensive use of eight natural resource management (NRM) clusters, which can be further divided into sub-clusters or combined into super-clusters.  
https://www.climatechangeinaustralia.gov.au/en/overview/methodology/nrm-regions/

The edges of the shapefiles available for download at that CCiA website don't exactly match the coastline of Australia (see figures below), so here we use NRM shapefiles provided by the Bureau of Meteorology which match the coastline.

```python
import geopandas as gp
from IPython.display import Image
import matplotlib.pyplot as plt
```

```python
#https://www.climatechangeinaustralia.gov.au/media/ccia/2.2/cms_page_media/238/NRM_CCIA_clusters.jpg
Image(filename='figures/NRM_CCIA_clusters.jpg')
```

```python
# https://www.climatechangeinaustralia.gov.au/media/ccia/2.2/cms_page_media/238/NRM_CCIA_sub_clusters.png
Image(filename='figures/NRM_CCIA_sub_clusters.png', width=500)
```

```python
# https://www.climatechangeinaustralia.gov.au/media/ccia/2.2/cms_page_media/238/NRM_CCIA_super_clusters.png
Image(filename='figures/NRM_CCIA_super_clusters.png', width=500)
```

```python
nrm_regions_file = 'NRM_regions_clipped_toAWRA/NRM_regions_clipped_toAWRA.shp'
```

```python
nrm_regions = gp.read_file(nrm_regions_file)
```

```python
nrm_regions = nrm_regions[0:15]
```

```python
nrm_regions = nrm_regions.drop(columns=['OBJECTID_1', 'OBJECTID', 'name', 'Shape_Leng', 'Shape_Area'])
nrm_regions = nrm_regions.rename(columns={'label': 'SubClusterName', 'code': 'SubClusterCode', 'C_code': 'ClusterCode', 'C_Label': 'ClusterName'})
```

```python
nrm_regions
```

```python
nrm_regions['SuperClusterName'] = [
    'Northern Australia',
    'Rangelands',
    'Northern Australia',
    'Northern Australia',
    'Eastern Australia',
    'Eastern Australia',
    'Southern Australia',
    'Southern Australia',
    'Southern Australia',
    'Southern Australia',
    'Southern Australia',
    'Southern Australia',
    'Southern Australia',
    'Eastern Australia',
    'Rangelands',
]
```

```python
nrm_regions['SuperClusterCode'] = [
    'NA',
    'R',
    'NA',
    'NA',
    'EA',
    'EA',
    'SA',
    'SA',
    'SA',
    'SA',
    'SA',
    'SA',
    'SA',
    'EA',
    'R',
]
```

```python
nrm_regions = nrm_regions[[
    'SubClusterName',
    'SubClusterCode',
    'ClusterName',
    'ClusterCode',
    'SuperClusterName',
    'SuperClusterCode',
    'geometry',
]]
```

```python
nrm_regions
```

```python
nrm_regions.plot(
    figsize=[5, 5],
    column='SubClusterName',
    cmap='tab20',
    legend=True,
    linewidth=0.25,
    edgecolor='black',
    legend_kwds={'loc': 'right', 'bbox_to_anchor': (2.0, 0.5)},
)
plt.title('NRM sub-clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
nrm_clusters = nrm_regions.dissolve(by='ClusterName', as_index=False)
nrm_clusters = nrm_clusters.drop(columns=['SubClusterName', 'SubClusterCode'])
nrm_clusters = nrm_clusters[['ClusterName', 'ClusterCode', 'SuperClusterName', 'SuperClusterCode', 'geometry']]
```

```python
nrm_clusters
```

```python
nrm_clusters.plot(
    figsize=[5, 5],
    column='ClusterName',
    cmap='tab10',
    legend=True,
    linewidth=0.25,
    edgecolor='black',
    legend_kwds={'loc': 'right', 'bbox_to_anchor': (1.9, 0.5)},
)
plt.title('NRM clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
nrm_super_clusters = nrm_regions.dissolve(by='SuperClusterName', as_index=False)
nrm_super_clusters = nrm_super_clusters.drop(columns=['SubClusterName', 'SubClusterCode', 'ClusterName', 'ClusterCode'])
nrm_super_clusters = nrm_super_clusters[['SuperClusterName', 'SuperClusterCode', 'geometry']]
```

```python
nrm_super_clusters
```

```python
nrm_clusters.plot(
    figsize=[5, 5],
    column='SuperClusterName',
    cmap='tab10',
    legend=True,
    linewidth=0.25,
    edgecolor='black',
    legend_kwds={'loc': 'right', 'bbox_to_anchor': (1.5, 0.5)},
)
plt.title('NRM super clusters')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
